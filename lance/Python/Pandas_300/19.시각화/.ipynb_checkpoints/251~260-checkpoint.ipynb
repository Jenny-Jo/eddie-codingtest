### 253. bar plot

iris 데이터에서 species별 sepal_length의 평균을 구한 후 이를 'bar' 타입으로 그려보세요.  
import seaborn as sns

iris = sns.load_dataset("iris")

```python
import seaborn as sns

iris = sns.load_dataset("iris")

iris.groupby('species')['sepal_length'].mean().plot(kind='bar')
```

```python
iris.groupby('species').mean()
```

### 254 DataFrame.plot.barh

iris 데이터에서 품종별 특징값들의 평균을 구한 후 이를 barh 형태로 그려보세요.

```python
iris.groupby('species').mean().plot(kind='barh')
```

```python
iris.groupby('species').mean().plot.barh()
```

### 255. DataFrame.T and barh

iris 데이터에서 품종별 특징값들의 평균을 구한 후 이를 transpose하고 barh 형태로 그려보세요.

```python
iris.groupby('species').mean().T.plot.barh()
```

### 256. pandas.Series.plot.pie

iris 데이터에서 품종별 데이터의 개수를 파이 차트 형태로 그려보세요.

```python
iris.groupby('species').count()['sepal_length'].plot.pie(autopct="%.2f%%")
```

### 257. seaborn.scatterplot

seaborn의 scatterplot 함수를 사용하여 iris 데이터에서 품종별 petal_length와 petal_width를 scatter 타입으로 그려보세요.

```python
import seaborn as sns

sns.scatterplot(
    x='petal_width',
    y='petal_length',
    hue='species',
    data=iris
)
```

### 258. seaborn.boxplot

seaborn 모듈의 boxplot 함수를 사용하여 다음과 같이 그려보세요.

```python
sns.boxplot( x='species',
    y='petal_length',
    data=iris)
```

### 259. seaborn.kdeplot

seaborn kdeplot 함수를 사용하여 iris 데이터로부터 다음과 같이 그려보세요.

```python
sns.kdeplot(x='petal_length',
           hue='species',
           data=iris)
```

### 260. seaborn.FacetGrid

seaborn의 FacetGrid를 사용해서 다음과 같이 품종별 히스토그램을 그려보세요.

```python
g = sns.FacetGrid(iris, col='species')
g.set(xlim=(0, 10))
g.map(sns.histplot, 'petal_length')
```
